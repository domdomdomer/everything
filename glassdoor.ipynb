```python
import pandas as pd
import numpy as np

df = pd.read_csv('Uncleaned_DS_jobs.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python

df = df.drop(['index'], axis = 1)
df.head()
```

```python

df = df.replace('-1', np.nan)
df.tail(10)
```

```python
df['Founded'].unique()
```

```python
df['Founded'] = df['Founded'].replace(-1, np.nan)
df['Founded'].unique()
```

```python
df.isna().sum()
```

```python
df['Size'].unique()
```

```python
df['Size'] = df['Size'].fillna(value = 'Unknown')
df['Size'].unique()
```

```python
df['Type of ownership'].unique()
```

```python
df['Type of ownership'] = df['Type of ownership'].fillna(value = 'Unknown')
df['Type of ownership'].unique()
```

```python
df['Sector'].unique()
```

```python
df['Industry'].unique()
```

```python
idx_sec = df[df['Sector'].isna()].index
idx_sec
```

```python
idx_ind = df[df['Industry'].isna()].index
idx_ind
```

```python
df['Industry'] = df['Industry'].fillna(value = 'Unknown')
df['Sector'] = df['Sector'].fillna(value = 'Unknown')
df.isna().sum()
```

```python
df['Revenue'].unique()
```

```python
df['Revenue'] = df['Revenue'].fillna(value = 'Unknown / Non-Applicable')
df['Revenue'].unique()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Job Description'], axis = 1)
df = df.drop(['Competitors'], axis = 1)
df.head()
```

```python
df = df.dropna(axis = 0)
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.reset_index(drop = True, inplace = True)
df.info()
```

```python
df[['Salary Estimate min', 'Salary Estimate max']] = df['Salary Estimate'].str.split('-', expand = True)
df.head()
```

```python
col_mov = df.pop('Salary Estimate min')
df.insert(2, 'Salary Estimate min', col_mov)
df.head()
```

```python
col_mov2 = df.pop('Salary Estimate max')
df.insert(3, 'Salary Estimate max', col_mov2)
df.head()
```

```python
df['Salary Estimate min'] = df['Salary Estimate min'].apply(lambda x: x.replace('$', '').strip())
df['Salary Estimate max'] = df['Salary Estimate max'].apply(lambda x: x.replace('$', '').strip())
df.head()
```

```python
df['Salary Estimate max'] = df['Salary Estimate max'].apply(lambda x: x.replace('(Glassdoor est.)', '').strip())
df.head()
```

```python
df['Salary Estimate max'].unique()
```

```python
df['Salary Estimate max'] = df['Salary Estimate max'].apply(lambda x: x.replace('(Employer est.)', '').strip())
df['Salary Estimate max'].unique()
```

```python
def conversor(valor):
  if 'K' in valor:
    k = round((float(valor.replace('K', '').strip()) * 1000), 2)
    return k
  else:
    return float(valor)
```

```python
df['Salary Estimate min'] = df['Salary Estimate min'].apply(conversor)
df['Salary Estimate max'] = df['Salary Estimate max'].apply(conversor)
df.head()
```

```python
df['Company Name'] = df['Company Name'].str.split('\n').str[0]
df.head()
```

```python
df['Company Name'].unique()
```

```python
df['Location'] = df['Location'].str.strip()
df.head()
```

```python
df['Location'].unique()
```

```python
idx = df[df['Location'].str.contains(',') == False].index
df.loc[idx]
```

```python
df.loc[idx, 'Location'].unique()
```

```python
df.loc[idx, 'Location'] = df.loc[idx, 'Location'].str.replace('United States', 'Remote')
df.loc[idx, 'Location'] = df.loc[idx, 'Location'].str.replace('New Jersey', 'Remote')
df.loc[idx, 'Location'] = df.loc[idx, 'Location'].str.replace('Utah', 'Remote')
df.loc[idx, 'Location'] = df.loc[idx, 'Location'].str.replace('Texas', 'Remote')
df.loc[idx, 'Location'] = df.loc[idx, 'Location'].str.replace('California', 'Remote')
df.loc[idx]
```

```python
df.loc[idx, 'Location'] = df.loc[idx, 'Location'] + ',' + df.loc[idx, 'Headquarters'].str.split(',').str[1]
df.loc[idx]
```

```python
df[['City', 'State']] = df['Location'].str.split(',', n = 1, expand = True)
df.head()
```

```python
df['State'].unique()
```

```python
idx_anne = df[df['State'].str.contains('Anne Arundel') == True].index
idx_anne
```

```python
df.loc[idx_anne, 'State'] = df.loc[idx_anne, 'State'].str.replace(' Anne Arundel, ', '')
df.loc[idx_anne]
```

```python

col_mov3 = df.pop('City')
df.insert(7, 'City', col_mov3)
df.head()
```

```python
col_mov4 = df.pop('State')
df.insert(8, 'State', col_mov4)
df.head()
```

```python
df = df.drop(['Location'], axis = 1)
df.head()
```

```python

df['Founded'] = df['Founded'].astype(int)
```

```python
df['Salary Estimate mean'] = (df['Salary Estimate max'] + df['Salary Estimate min']) / 2
df.head()
```

```python
col_mov5 = df.pop('Salary Estimate mean')
df.insert(4, 'Salary Estimate mean', col_mov5)
df.head()
```

```python
df.info()
```

```python
df.to_csv('clean_glassdoor_data.csv', index = False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
fig = plt.figure(figsize = (8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

sns.kdeplot(df['Rating'], fill = True)
plt.title('Distribuição do Rating das empresas', pad = 10)
eixo.set_xlabel('Ratings')
eixo.set_ylabel('Densidade')
plt.show()
```

```python
fig = plt.figure(figsize = (8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

sns.kdeplot(df['Salary Estimate mean'], fill = True, color = 'darkslateblue')
plt.title('Distribuição de média salarial')
eixo.set_xlabel('Salários')
eixo.set_ylabel('Densidade')
plt.show()
```

```python
fig = plt.figure(figsize = (8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

sns.kdeplot(df['Founded'], fill = True, color = 'forestgreen')
plt.title('Distribuição do ano de fundação das empresas', pad = 10)
eixo.set_xlabel('Ano de fundação')
eixo.set_ylabel('Densidade')
plt.show()
```

```python
cont_state = df['State'].value_counts()
top10s = cont_state.head(10)

fig = plt.figure(figsize = (8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

viz = sns.countplot(x='State', data = df[df['State'].isin(top10s.index)])
plt.title('Estados com mais pessoas empregadas em Data Science', pad = 10)
eixo.set_xlabel('Estados')
eixo.set_ylabel('Número de profissionais')
plt.show()
```

```python
cont_sec = df['Sector'].value_counts()
top10 = cont_sec.head(10)

fig = plt.figure(figsize = (8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

viz = sns.countplot(x='Sector', data = df[df['Sector'].isin(top10.index)], color = 'tab:blue')
plt.title('Setores que mais empregaram profissionais de Data Science', pad = 10)
eixo.set_xlabel('Setor')
eixo.set_ylabel('Profissionais empregados')
viz.set_xticklabels(viz.get_xticklabels(), rotation = 30, ha = 'right')
plt.show()
```
